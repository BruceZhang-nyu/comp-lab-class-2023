```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y=np.loadtxt('../Figures/potential.xvg',skiprows=24,unpack=True)
plt.plot(x,y,label='Potential')
plt.xlabel('Energy Minimization Step')
plt.ylabel('Potential Energy(kj mol$^{-1}$)')
plt.title('Erergy Minimization(1AKI, Steepest Descent)')
plt.legend()
plt.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
```

```python
x,y=np.loadtxt('../Figures/temperature.xvg',skiprows=24,unpack=True)
plt.plot(x,y,label='Tmepeture')
plt.title('Temperature (1AKI, NVT Equilibration)')
plt.xlabel('Times (ps)')
plt.ylabel('Temperature(K)')
```

```python
x,y=np.loadtxt('../Figures/pressure.xvg',skiprows=24,unpack=True)

def mean_10ns(x,y):
    y_mean=[]
    x_mean=[]
    for i in range(5,96):
        y_mean.append(np.mean(y[i-5:i+4]))
        x_mean.append(i-0.5)
    return x_mean,y_mean
x_mean,y_mean=mean_10ns(x,y)
plt.plot(x,y)
plt.plot(x_mean,y_mean,label='10ps Running Avg',color='r')
plt.title('Pressure(1AKI, NPT Equilibration)')
plt.xlabel('Time(ps)')
plt.ylabel('Pressure(bar)')
plt.legend()

```

```python
x,y=np.loadtxt('../Figures/density.xvg',skiprows=24,unpack=True)
x_mean,y_mean=mean_10ns(x,y)
plt.plot(x,y)
plt.plot(x_mean,y_mean,label='10ps Running Avg',color='r')
plt.title('Density(1AKI, NPT Equilibration)')
plt.xlabel('Time(ps)')
plt.ylabel('Density(kg m$^{-3}$)')
plt.legend()
```

```python
x_equi,y_equi=np.loadtxt('../Figures/rmsd.xvg',skiprows=24,unpack=True)
x_crys,y_crys=np.loadtxt('../Figures/rmsd_xtal.xvg',skiprows=24,unpack=True)
plt.plot(x_equi[::50],y_equi[::50],label='Ref:Equilibrated',color='black')
plt.plot(x_crys[::50],y_crys[::50],label='Ref:Crystal',color='r')
plt.tight_layout()
plt.title('RMSD (1AKI, Backbone)')
plt.xlabel('Time(ns)')
plt.ylabel('RMSD(nm)')
plt.legend(loc=4)
```
